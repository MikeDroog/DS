#### Model trainen met een DecisionTree
Het volgende plaatje laat het proces zien van een model trainen en testen. Neem de volgende stappen: 
1. importeer van sklearn de decisiontreeclassifier. Waarom gebruiken we deze?
2. fit de trainingset op de testset waarvan het resultaat bekend is om de classificatiewaardes te verkrijgen
3. Laat je bevindingen weten

<img style="float: left;" src="titanic/data_and_model.png">  

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
complete['Embarked_Numeric'] = le.fit_transform(complete['Embarked'])
complete['Sex_Numeric'] = le.fit_transform(complete['Sex'])
complete.info()
```

```python
complete['Title'] = complete.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
complete['Title'].unique()
complete['Title'] = complete['Title'].replace(['Capt', 'Col',
                                          'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer'], 'UncommonMale')
complete['Title'] = complete['Title'].replace(['Lady', 'Countess', 'Dona'], 'UncommonFemale')
complete['Title'] = complete['Title'].replace('Mlle', 'Miss')
complete['Title'] = complete['Title'].replace('Ms', 'Miss')
complete['Title'] = complete['Title'].replace('Mme', 'Mrs')
complete['Title'] = complete['Title'].map( {'UncommonMale': 0, 'UncommonFemale': 1, 'Mr': 2, 'Mrs': 3, 'Miss': 4, 'Master': 5} ).astype(int)
```

```python
if 'Ticket' in complete.columns:
    complete= complete.drop(['Ticket'],axis=1)
if 'Sex' in complete.columns:
    complete= complete.drop(['Sex'],axis=1)
if 'Age' in complete.columns:
    complete= complete.drop(['Age'],axis=1)
if 'Fare' in complete.columns:
    complete= complete.drop(['Fare'],axis=1)
if 'Embarked' in complete.columns:
    complete= complete.drop(['Embarked'],axis=1)
if 'Name' in complete.columns:
    complete= complete.drop(['Name'],axis=1)
```

```python
train = complete[complete['Survived'].notnull()]
test = complete[complete['Survived'].isnull()]
train_split, test_split = train_test_split(train, test_size=0.2)
print(len(train_split),len(test_split))

xtrain = train_split.drop(['Survived'], axis=1)
ytrain = train_split.Survived
xtest = test_split.drop(['Survived'], axis=1)
ytest = test_split.Survived
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(xtrain,ytrain)
```

```python
ypred = decision_tree.predict(xtest)
acc_decision_tree = round(decision_tree.score(xtest,ytest) * 100,2)
dec_par = decision_tree.get_params()
print(acc_decision_tree,"%")
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(decision_tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
