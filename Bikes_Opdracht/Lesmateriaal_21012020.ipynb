#### Embarked (Opdracht: omzetten naar numeriek)  

```python
# Code
# 3 waardes naar een OneHotEncoder of een waarde geven maar als er geen rangschikking is binnen categorische data 
# dan is het toekennen van een waarde vaak misleidend voor een model want die interpreteert een getal als dat er 
# een hierarchie zit in de data terwijl die er helemaal niet is! 

complete['Embarked_Numeric'] = complete['Embarked'].m ap( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

#### Name (Opdracht: Wat voor bruikbare informatie kan je hieruit halen? Kan deze feature ook waardes van andere features voorspellen?)    

```python
# Code
# Nieuwe feature maken met alleen de aanhef en die ook als een OneHotEncoder omvormen

# Hou alleen de titel over
complete['Title'] = complete.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
complete['Title'].unique()
complete['Title'] = complete['Title'].replace(['Capt', 'Col',
                                           'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer'], 'UncommonMale')
complete['Title'] = complete['Title'].replace(['Lady', 'Countess', 'Dona'], 'UncommonFemale')
complete['Title'] = complete['Title'].replace('Mlle', 'Miss')
complete['Title'] = complete['Title'].replace('Ms', 'Miss')
complete['Title'] = complete['Title'].replace('Mme', 'Mrs')
complete['Title'] = complete['Title'].map( {'UncommonMale': 0, 'UncommonFemale': 1, 'Mr': 2, 'Mrs': 3, 'Miss': 4, 'Master': 5} ).astype(int)
```

#### Sex (Opdracht: Zet om naar 1 en 0)

```python
# Code
# Zet man en vrouw om in 0 en 1

if complete['Sex'].values.dtype != int:
    print("Sex to numeric")
    complete['Sex'] = complete['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

#### Ticket (Opdracht: Naar numeriek en anders: wat is een goed plan?)  

```python
# Code
# Maak een afweging, weggooien of alleen de cijfers onthouden als numeric?

if'Ticket' in complete.columns:
    print("Ticket to numeric")
    complete = complete.drop(['Ticket'], axis=1)

```
